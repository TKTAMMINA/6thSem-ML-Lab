```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler, OneHotEncoder

# Load datasets
adult_df = pd.read_csv('/content/adult.csv')
diabetes_df = pd.read_csv('/content/Dataset of Diabetes .csv')

# Function to handle missing values
def handle_missing_values(df):
    missing_cols = df.columns[df.isnull().any()].tolist()
    print(f"Columns with missing values: {missing_cols}")
    df.fillna(df.median(numeric_only=True), inplace=True)  # Fill numerical NaN with median
    df.fillna(df.mode().iloc[0], inplace=True)  # Fill categorical NaN with mode
    return df

# Function to encode categorical data
def encode_categorical_data(df):
    categorical_columns = df.select_dtypes(include=['object']).columns.tolist()
    print(f"Categorical columns identified: {categorical_columns}")
    if categorical_columns:
        encoder = OneHotEncoder(sparse_output=False, drop='first', handle_unknown='ignore')
        encoded_arr = encoder.fit_transform(df[categorical_columns])
        encoded_df = pd.DataFrame(encoded_arr, columns=encoder.get_feature_names_out(categorical_columns))
        df = df.drop(columns=categorical_columns).reset_index(drop=True)
        df = pd.concat([df, encoded_df], axis=1)
    return df

# Function to handle outliers
def cap_outliers(df):
    for col in df.select_dtypes(include=[np.number]).columns:
        lower_bound = df[col].quantile(0.05)
        upper_bound = df[col].quantile(0.95)
        df[col] = np.clip(df[col], lower_bound, upper_bound)
    return df

# Function to apply scaling
def apply_scaling(df):
    scaler_minmax = MinMaxScaler()
    scaler_standard = StandardScaler()
    numeric_cols = df.select_dtypes(include=[np.number]).columns
    df[[col + '_minmax' for col in numeric_cols]] = scaler_minmax.fit_transform(df[numeric_cols])
    df[[col + '_standard' for col in numeric_cols]] = scaler_standard.fit_transform(df[numeric_cols])
    return df

# Function to apply scaling
def apply_scaling(df):
    scaler_minmax = MinMaxScaler()
    scaler_standard = StandardScaler()

    numeric_cols = df.select_dtypes(include=[np.number]).columns
    minmax_scaled = scaler_minmax.fit_transform(df[numeric_cols])
    standard_scaled = scaler_standard.fit_transform(df[numeric_cols])

    minmax_df = pd.DataFrame(minmax_scaled, columns=[col + '_minmax' for col in numeric_cols])
    standard_df = pd.DataFrame(standard_scaled, columns=[col + '_standard' for col in numeric_cols])

    df = df.reset_index(drop=True)  # Reset index before concatenation to avoid misalignment
    return pd.concat([df, minmax_df, standard_df], axis=1)


# Process the datasets
adult_df = handle_missing_values(adult_df)
adult_df = encode_categorical_data(adult_df)
adult_df = cap_outliers(adult_df)
adult_df = apply_scaling(adult_df)

print("Processed Adult Dataset:")
print(adult_df.head())

diabetes_df = handle_missing_values(diabetes_df)
diabetes_df = encode_categorical_data(diabetes_df)
diabetes_df = cap_outliers(diabetes_df)
diabetes_df = apply_scaling(diabetes_df)

print("Processed Diabetes Dataset:")
print(diabetes_df.head())
```
